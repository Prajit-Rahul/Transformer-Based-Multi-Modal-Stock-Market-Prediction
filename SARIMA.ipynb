```python
import numpy as np  # linear algebra
import pandas as pd  # data processing, CSV file I/O (e.g. pd.read_csv)
import yfinance as yf  # for fetching stock data
import matplotlib.pyplot as plt
```

```python
def convert_data_type(df, time_cols=[], float_cols=[], cat_cols=[]):
    for col in time_cols:
        df[col] = pd.to_datetime(df[col], format='%Y-%m-%d')
    for col in float_cols:
        df[col] = df[col].astype(float)
    for col in cat_cols:
        df[col] = df[col].astype('category')
    return df

# Define the stock tickers and the date range
tickers = ['ADBE', 'BX', 'MET', 'DLR', 'TJX', 'TT', 'BN', 'MA', 'CPRT', 'COST', 'NOC', 'UNP', 'BKNG', 'MDLZ', 'ADI', 'WM', 'TDG', 'CI', 'SYK', 'MU', 'ABBV', 'ABNB', 'PEP', 'NOW', 'LLY', 'ISRG', 'CSCO', 'EQIX', 'HCA', 'GILD', 'VRTX', 'APH', 'VZ', 'CL', 'MSI', 'FCX', 'CDNS', 'PSX', 'BAC', 'UAL', 'CHTR', 'MDT', 'BK', 'MS', 'ROP', 'CTAS', 'MSFT', 'ZTS', 'FDX', 'RGNX', 'AMAT', 'AON', 'AZO', 'AXP', 'UBER', 'CME', 'PH', 'AMP', 'CVX', 'XOM', 'ICE', 'PCG', 'NXPI', 'KLAC', 'NVDA', 'ACN', 'SCHW', 'SQ', 'CMG', 'SHW', 'SPG', 'AMD', 'O', 'URI', 'RTX', 'PNC', 'SBUX', 'ORCL', 'TXN', 'MMM', 'TMO', 'INTU', 'DHI', 'INTC', 'AMZN', 'MNST', 'UPS', 'NEE', 'ELV', 'EMR', 'ECL', 'WELL', 'AMT', 'GE', 'PCAR', 'MCD', 'PM', 'ADSK', 'CAT', 'GS', 'GD', 'FTNT', 'PYPL', 'BDX', 'KMI', 'FICO', 'D', 'DE', 'SRE', 'TSLA', 'RCL', 'ANET', 'ABT', 'CVS', 'META', 'SO', 'COF', 'PAYX', 'COP', 'DHR', 'MCK', 'CRWD', 'HD', 'ORLY', 'WFC', 'NEM', 'FANG', 'MCO', 'DIS', 'PG', 'ALL', 'T', 'C', 'ETN', 'GWW', 'NRG', 'NSC', 'SPGI', 'NKE', 'V', 'KO', 'PANW', 'MAR', 'FI', 'GM', 'DELL', 'IBM', 'PLD', 'BSX', 'USB', 'NFLX', 'EOG', 'LIN', 'ITW', 'BMY', 'PSA', 'GEV', 'PLTR', 'AEP', 'TRV', 'CB', 'TMUS', 'JNJ', 'CRM', 'AAPL', 'PGR', 'BLK', 'WMB', 'AIG', 'TGT', 'SLB', 'SON', 'APD', 'CEG', 'HLT', 'WMT', 'LRCX', 'LMT', 'SNPS', 'GOOG', 'OKE', 'MO', 'REGN', 'AVGO', 'TFC', 'CSX', 'BA', 'VFC', 'AFL', 'AMGN', 'IMVT', 'UNH', 'QCOM', 'PFE', 'CARR', 'DUK', 'JCI', 'VST', 'ADP', 'RSG']

start_date = "2024-02-03"
end_date = "2024-11-03"
```

```python
# Download historical data from yfinance
df = yf.download(tickers, start=start_date, end=end_date, group_by='ticker')
```

```python
df = df.stack(level=0).reset_index()
df.columns = ['time', 'ticker', 'open', 'high', 'low', 'close', 'adjclose', 'volume']
df = convert_data_type(df, time_cols=['time'], float_cols=['open', 'high', 'low', 'close', 'volume'], cat_cols=['ticker'])

# Set the time as index and sort values by time
df = df.set_index('time', drop=True).reset_index()
df = df.groupby('ticker').apply(lambda x: x.sort_values('time')).reset_index(drop=True)
```

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt


# Define cutoff dates for train, validation, and test sets
test_cutoff = df.index.max() - pd.Timedelta(days=30)
validation_cutoff = test_cutoff - pd.Timedelta(days=30)

# Reset the index of the original dataframe
df_reset = df.reset_index()

# Create exogenous variables
df_exog = pd.get_dummies(df_reset['ticker'], drop_first=True)
df_exog['time'] = df_reset['time']
df_exog.set_index('time', inplace=True)

# Now split the data again
train_exog = df_exog[df_exog.index <= validation_cutoff]
val_exog = df_exog[(df_exog.index > validation_cutoff) & (df_exog.index <= test_cutoff)]
test_exog = df_exog[df_exog.index > test_cutoff]

# Split data
train_data = df[df.index <= validation_cutoff]
train_exog = df_exog.loc[train_data.index]

val_data = df[(df.index > validation_cutoff) & (df.index <= test_cutoff)]
val_exog = df_exog.loc[val_data.index]

test_data = df[df.index > test_cutoff]
test_exog = df_exog.loc[test_data.index]

# Remove duplicate timestamps in each subset
# Remove duplicate timestamps in each subset
# Remove duplicate timestamps in each subset
train_data = train_data[~train_data.index.duplicated(keep='first')]
val_data = val_data[~val_data.index.duplicated(keep='first')]
test_data = test_data[~test_data.index.duplicated(keep='first')]

train_exog = train_exog.loc[train_data.index]
val_exog = val_exog.loc[val_data.index]
test_exog = test_exog.loc[test_data.index]

# Verify shapes
print(train_data.shape)
print(train_exog.shape)

# Fit SARIMA model with exogenous variable (ticker one-hot encoding)
model = SARIMAX(train_data['close'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 7), exog=train_exog)
sarima_result = model.fit()

# Forecast for the validation period
val_forecast = sarima_result.get_forecast(steps=len(val_data), exog=val_exog)
val_data['forecast'] = val_forecast.predicted_mean


```

```python
print(train_data.head())
print(train_exog.head())
print(train_data.shape)
print(train_exog.shape)
```

```python
# Calculate validation errors
val_mae = mean_absolute_error(val_data['close'], val_data['forecast'])
val_mse = mean_squared_error(val_data['close'], val_data['forecast'])
print(f"Validation MAE: {val_mae}, Validation MSE: {val_mse}")

# Forecast for the test period
test_forecast = sarima_result.get_forecast(steps=len(test_data), exog=test_exog)
test_data['forecast'] = test_forecast.predicted_mean

# Calculate test errors
test_mae = mean_absolute_error(test_data['close'], test_data['forecast'])
test_mse = mean_squared_error(test_data['close'], test_data['forecast'])
print(f"Test MAE: {test_mae}, Test MSE: {test_mse}")

```

```python
# Plot the results
plt.figure(figsize=(14, 7))
plt.plot(train_data['close'], label='Training Data')
plt.plot(val_data['close'], label='Validation Data', color='orange')
plt.plot(val_data['forecast'], label='Validation Forecast', color='green')
plt.plot(test_data['close'], label='Test Data', color='red')
plt.plot(test_data['forecast'], label='Test Forecast', color='purple')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title("SARIMA Model with Ticker as a Categorical Variable")
plt.legend()
plt.show()
```

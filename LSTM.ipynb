```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
import math

# Define the stock tickers and date range
tickers = ['AAPL', 'MSFT', 'GOOGL']
start_date = "2018-06-01"
end_date = "2024-06-30"

# Fetch stock data for multiple tickers
df = yf.download(tickers, start=start_date, end=end_date, group_by='ticker')

# Convert the multi-index DataFrame to a single DataFrame
df = df.stack(level=0).reset_index()
df.columns = ['time', 'ticker', 'open', 'high', 'low', 'close', 'adjclose', 'volume']

# Function to convert data types
def convert_data_type(df, time_cols, float_cols, cat_cols):
    for col in time_cols:
        df[col] = pd.to_datetime(df[col])
    for col in float_cols:
        df[col] = df[col].astype(float)
    for col in cat_cols:
        df[col] = df[col].astype('category')
    return df

# Apply the data type conversion
df = convert_data_type(df, time_cols=['time'], float_cols=['open', 'high', 'low', 'close', 'volume'], cat_cols=['ticker'])

# Set the time as index and sort values by time for each ticker
df = df.set_index('time', drop=True).reset_index()
df = df.groupby('ticker').apply(lambda x: x.sort_values('time')).reset_index(drop=True)

# Preprocess and scale the data
def preprocess_data(ticker_df):
    ticker_df = ticker_df[['close', 'volume', 'open', 'high', 'low']]  # Include more features for robust learning
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(ticker_df)
    return scaled_data, scaler

# Prepare data for LSTM (using 60 timesteps)
def create_dataset(dataset, time_step=60):
    X, y = [], []
    for i in range(len(dataset) - time_step - 1):
        X.append(dataset[i:(i + time_step), :])  # Using all features
        y.append(dataset[i + time_step, 0])  # Predicting 'close' price
    return np.array(X), np.array(y)

# Iterate through each ticker and train LSTM
for ticker in tickers:
    print(f"\nProcessing ticker: {ticker}")
    ticker_df = df[df['ticker'] == ticker]
    
    # Preprocess data
    scaled_data, scaler = preprocess_data(ticker_df)
    
    # Create dataset
    time_step = 60
    X, y = create_dataset(scaled_data, time_step)
    
    # Reshape the data to [samples, timesteps, features]
    X = X.reshape(X.shape[0], X.shape[1], X.shape[2])
    
    # Split into training, validation, and test sets
    train_size = int(len(X) * 0.7)
    val_size = int(len(X) * 0.2)
    X_train, X_val, X_test = X[:train_size], X[train_size:train_size + val_size], X[train_size + val_size:]
    y_train, y_val, y_test = y[:train_size], y[train_size:train_size + val_size], y[train_size + val_size:]
    
    # Build the LSTM model
    model = Sequential()
    model.add(LSTM(units=64, return_sequences=True, input_shape=(time_step, X.shape[2])))
    model.add(Dropout(0.3))  # Dropout increased to 30% to avoid overfitting
    model.add(LSTM(units=64, return_sequences=False))
    model.add(Dropout(0.3))
    model.add(Dense(units=1))  # Output layer predicting the closing price
    
    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')
    
    # Early stopping to prevent overfitting
    early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    
    # Train the model with validation set
    model.fit(X_train, y_train, epochs=50, batch_size=64, validation_data=(X_val, y_val), callbacks=[early_stop])
    
    # Make predictions on the test set
    predictions = model.predict(X_test)
    predictions = scaler.inverse_transform(np.concatenate((predictions, np.zeros((len(predictions), X.shape[2] - 1))), axis=1))[:, 0]  # Inverse scale only the 'close' column

    # Inverse transform the test data for comparison
    y_test_rescaled = scaler.inverse_transform(np.concatenate((y_test.reshape(-1, 1), np.zeros((len(y_test), X.shape[2] - 1))), axis=1))[:, 0]

    # Compute MAE
    mae = mean_absolute_error(y_test_rescaled, predictions)
    
    # Compute RMSE
    rmse = math.sqrt(mean_squared_error(y_test_rescaled, predictions))
    
    # Compute R-Squared
    r2 = r2_score(y_test_rescaled, predictions)
    
    # Print results for the ticker
    print(f"Mean Absolute Error for {ticker}: {mae}")
    print(f"Root Mean Squared Error for {ticker}: {rmse}")
    print(f"R-Squared for {ticker}: {r2}")
    
    # Plot the predictions and test data
    plt.figure(figsize=(12, 6))
    plt.plot(y_test_rescaled, label='Actual Price')
    plt.plot(predictions, label='Predicted Price')
    plt.title(f'{ticker} Price Prediction')
    plt.xlabel('Time')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
import math

# Define the stock tickers and date range
tickers = ['AAPL', 'MSFT', 'GOOGL']
start_date = "2018-06-01"
end_date = "2024-06-30"

# Fetch stock data for multiple tickers
df = yf.download(tickers, start=start_date, end=end_date, group_by='ticker')

# Convert the multi-index DataFrame to a single DataFrame
df = df.stack(level=0).reset_index()
df.columns = ['time', 'ticker', 'open', 'high', 'low', 'close', 'adjclose', 'volume']

# Function to convert data types
def convert_data_type(df, time_cols, float_cols, cat_cols):
    for col in time_cols:
        df[col] = pd.to_datetime(df[col])
    for col in float_cols:
        df[col] = df[col].astype(float)
    for col in cat_cols:
        df[col] = df[col].astype('category')
    return df

# Apply the data type conversion
df = convert_data_type(df, time_cols=['time'], float_cols=['open', 'high', 'low', 'close', 'volume'], cat_cols=['ticker'])

# Set the time as index and sort values by time for each ticker
df = df.set_index('time', drop=True).reset_index()
df = df.groupby('ticker').apply(lambda x: x.sort_values('time')).reset_index(drop=True)

# Preprocess and scale the data
def preprocess_data(ticker_df):
    ticker_df = ticker_df[['close', 'volume', 'open', 'high', 'low']]  # Include more features for robust learning
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(ticker_df)
    return scaled_data, scaler

# Prepare data for LSTM (using 60 timesteps)
def create_dataset(dataset, time_step=60):
    X, y = [], []
    for i in range(len(dataset) - time_step - 1):
        X.append(dataset[i:(i + time_step), :])  # Using all features
        y.append(dataset[i + time_step, 0])  # Predicting 'close' price
    return np.array(X), np.array(y)

# Iterate through each ticker and train LSTM
for ticker in tickers:
    print(f"\nProcessing ticker: {ticker}")
    ticker_df = df[df['ticker'] == ticker]
    
    # Preprocess data
    scaled_data, scaler = preprocess_data(ticker_df)
    
    # Create dataset
    time_step = 60
    X, y = create_dataset(scaled_data, time_step)
    
    # Reshape the data to [samples, timesteps, features]
    X = X.reshape(X.shape[0], X.shape[1], X.shape[2])
    
    # Split into training, validation, and test sets
    train_size = int(len(X) * 0.7)
    val_size = int(len(X) * 0.2)
    X_train, X_val, X_test = X[:train_size], X[train_size:train_size + val_size], X[train_size + val_size:]
    y_train, y_val, y_test = y[:train_size], y[train_size:train_size + val_size], y[train_size + val_size:]
    
    # Build a more complex LSTM model
    model = Sequential()

    # First LSTM layer (Bidirectional)
    model.add(Bidirectional(LSTM(units=128, return_sequences=True, input_shape=(time_step, X.shape[2]))))
    model.add(Dropout(0.3))  # Dropout to avoid overfitting
    model.add(BatchNormalization())  # Batch normalization to speed up training

    # Second LSTM layer (Bidirectional)
    model.add(Bidirectional(LSTM(units=128, return_sequences=True)))
    model.add(Dropout(0.3))
    model.add(BatchNormalization())

    # Third LSTM layer (Unidirectional)
    model.add(LSTM(units=64, return_sequences=False))  # Final LSTM layer with no return sequences
    model.add(Dropout(0.3))
    
    # Fully connected output layer
    model.add(Dense(units=1))  # Predicting closing price
    
    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')
    
    # Early stopping to prevent overfitting
    early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    
    # Train the model with validation set
    model.fit(X_train, y_train, epochs=100, batch_size=64, validation_data=(X_val, y_val), callbacks=[early_stop])
    
    # Make predictions on the test set
    predictions = model.predict(X_test)
    predictions = scaler.inverse_transform(np.concatenate((predictions, np.zeros((len(predictions), X.shape[2] - 1))), axis=1))[:, 0]  # Inverse scale only the 'close' column

    # Inverse transform the test data for comparison
    y_test_rescaled = scaler.inverse_transform(np.concatenate((y_test.reshape(-1, 1), np.zeros((len(y_test), X.shape[2] - 1))), axis=1))[:, 0]

    # Compute MAE
    mae = mean_absolute_error(y_test_rescaled, predictions)
    
    # Compute RMSE
    rmse = math.sqrt(mean_squared_error(y_test_rescaled, predictions))
    
    # Compute R-Squared
    r2 = r2_score(y_test_rescaled, predictions)
    
    # Print results for the ticker
    print(f"Mean Absolute Error for {ticker}: {mae}")
    print(f"Root Mean Squared Error for {ticker}: {rmse}")
    print(f"R-Squared for {ticker}: {r2}")
    
    # Plot the predictions and test data
    plt.figure(figsize=(12, 6))
    plt.plot(y_test_rescaled, label='Actual Price')
    plt.plot(predictions, label='Predicted Price')
    plt.title(f'{ticker} Price Prediction')
    plt.xlabel('Time')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.show()
```
